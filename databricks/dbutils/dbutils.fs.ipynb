### AWS S3 bucket

https://docs.aws.amazon.com/cli/latest/userguide/getting-started-quickstart.html

````
access_key = dbutils.secrets.get(scope = "aws-secrets", key = "access_key")
secret_key = dbutils.secrets.get(scope = "aws-secrets", key = "secret_key")

aws_bucket_name = "ocr-extraction-olonok"
mount_name ="ocr"

dbutils.fs.mount(f"s3a://{access_key}:{secret_key}@{aws_bucket_name}", f"/mnt/{mount_name}")
display(dbutils.fs.ls(f"/mnt/{mount_name}"))
````

````
display(dbutils.fs.ls(f"/mnt/{mount_name}/input"))
````

````
dbutils.fs.unmount(f"/mnt/{mount_name}")
````

### AZURE Blob
https://docs.databricks.com/en/dbfs/mounts.html

https://docs.databricks.com/en/connect/storage/aad-storage-service-principal.html


````
tenant_id = dbutils.secrets.get(scope = "aws-secrets", key = "azure-directory-id")
configs = {"fs.azure.account.auth.type": "OAuth",
          "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
          "fs.azure.account.oauth2.client.id": dbutils.secrets.get(scope = "aws-secrets", key = "azure-client-id"),
          "fs.azure.account.oauth2.client.secret": dbutils.secrets.get(scope = "aws-secrets", key = "azure-client-secret"),
          "fs.azure.account.oauth2.client.endpoint": f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"}

# Optionally, you can add <directory-name> to the source URI of your mount point.
dbutils.fs.mount(
  source = "abfss://test-dbs@olonokml9513337539.dfs.core.windows.net/",
  mount_point = "/mnt/azure",
  extra_configs = configs)
````

````
display(dbutils.fs.ls("/mnt/azure"))
````

````
dbutils.fs.unmount("/mnt/azure")
````
